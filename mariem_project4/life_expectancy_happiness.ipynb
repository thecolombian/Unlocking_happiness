```python
import pandas as pd



life_expectancy_df=pd.read_csv('life-expectancy-vs-healthcare-expenditure-2000-2021.csv')

life_expectancy_df
```

```python
life_expectancy_df = life_expectancy_df.dropna(subset=['Life expectancy - Sex: all - Age: 0 - Variant: estimates'])
life_expectancy_df
```

```python
life_expectancy_df = life_expectancy_df.dropna(subset=['Current health expenditure per capita, PPP (current international $)'])
life_expectancy_df
```

```python
unique_entities = life_expectancy_df['Entity'].unique()
print(unique_entities)
```

```python
counts_entities = life_expectancy_df['Entity'].value_counts()
print(counts_entities)
```

```python

life_expectancy_df = life_expectancy_df.rename(columns={
    'Current health expenditure per capita, PPP (current international $)': 'Health Expenditure Per Capita', 
    'Life expectancy - Sex: all - Age: 0 - Variant: estimates': "Life Expectancy"})
life_expectancy_df
```

```python
unique_years = life_expectancy_df['Year'].unique()
print(unique_years)
```

```python
life_expectancy_df = life_expectancy_df.rename(columns={
    'Entity': 'Country'})
life_expectancy_df
```

```python
counts_entities = life_expectancy_df['Country'].value_counts()
print(counts_entities)
```

```python
life_expectancy_df.to_csv('life_expectancy_cleaned.csv', index=False)
```

```python
happiness_df=pd.read_csv('filtered_happiness_data.csv')
print(life_expectancy_df.columns)
print(happiness_df.columns)
```

```python

merged_df = pd.merge(life_expectancy_df, happiness_df, on=['Year', 'Country'], how='outer')

# Check for duplicate columns (like 'Code') and handle them
# Identify columns that are duplicates
duplicate_columns = [col for col in merged_df.columns if 'Code' in col]

# If there are duplicate 'Code' columns, drop one
if len(duplicate_columns) > 1:
    merged_df = merged_df.drop(columns=[duplicate_columns[1]])  # Drop the second 'Code' column

# Optionally, rename the remaining 'Code' column if needed
merged_df.rename(columns={duplicate_columns[0]: 'Code'}, inplace=True)

# Display the merged DataFrame
print(merged_df)
```

```python
# Perform an outer join
merged_df = pd.merge(life_expectancy_df,happiness_df, on=['Year', 'Country'], how='inner')

# Check for duplicate columns (e.g., 'Code') and handle them
# If 'Code' exists in both DataFrames, it will result in 'Code_x' and 'Code_y'
if 'Code' in merged_df.columns:
    # Drop one of the duplicate columns if they are the same or if one is sufficient
    merged_df = merged_df.drop(columns=['Code_y'])  # Keep 'Code_x'

    # Optionally rename 'Code_x' to 'Code'
    merged_df.rename(columns={'Code_x': 'Code'}, inplace=True)

# Display the merged DataFrame
merged_df
```

```python
# Drop the 'Code_y' column
merged_df.drop(columns=['Code_y'], inplace=True)

# Rename 'Code_x' to 'Code'
merged_df.rename(columns={'Code_x': 'Code'}, inplace=True)

# Display the updated merged DataFrame
merged_df
```

```python
merged_df.to_csv('life_happiness_merged.csv', index=False)
```
